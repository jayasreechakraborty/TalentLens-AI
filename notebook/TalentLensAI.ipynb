```python
columns = [
    "title",
    "companyName",
    "location",
    "experience",
    "salary",
    "jobDescription",
    "tagsAndSkills",
    "jobUploaded",
    "minimumSalary",
    "maximumSalary",
    "minimumExperience",
    "maximumExperience"
]

df = df[columns]
```

```python
df["title"] = df["title"].str.strip()
```

```python
tech_keywords = [
    "Software Engineer",
    "Software Developer",
    "Java Developer",
    "Python Developer",
    "Data Analyst",
    "Data Scientist",
    "Data Engineer",
    "Business Analyst",
    "Machine Learning Engineer",
    "AI Engineer",
    "DevOps Engineer",
    "Cloud Engineer",
    "QA Engineer",
    "Automation Test Engineer",
    "Full Stack Developer",
    "Frontend Developer",
    "Backend Developer",
    "Web Developer",
    "SAP Consultant",
    "BI Developer"
]
```

```python
pattern = "|".join(tech_keywords)
```

```python
tech_df = df[
    df["title"].str.contains(pattern, case=False, na=False)
]
```

```python
exclude = [
    "Sales",
    "Marketing",
    "Recruiter",
    "HR",
    "Business Development",
    "Customer",
    "Finance",
    "Mechanical",
    "Civil",
    "Medical",
    "BPO",
    "Voice",
    "Non Voice",
    "Telecaller"
]
```

```python
exclude_pattern = "|".join(exclude)
```

```python
tech_df = tech_df[
    ~tech_df["title"].str.contains(
        exclude_pattern,
        case=False,
        na=False
    )
]
```

```python
tech_df.shape
```

```python
tech_df["title"].value_counts().head(20)
```

```python
salary_df = tech_df[
    (tech_df["minimumSalary"] > 0) &
    (tech_df["maximumSalary"] > 0)
]
```

```python
salary_df = salary_df.copy()

salary_df["averageSalary"] = (
    salary_df["minimumSalary"] +
    salary_df["maximumSalary"]
) / 2
```

```python
skills = (
    tech_df["tagsAndSkills"]
    .dropna()
    .str.split(",")
    .explode()
    .str.strip()
)
```

```python
skills = skills.replace({
    "python":"Python",
    "sql":"SQL",
    "javascript":"JavaScript",
    "css":"CSS",
    "html":"HTML",
    "aws":"AWS",
    "react.js":"React",
    "spring boot":"Spring Boot",
    "hibernate":"Hibernate",
    "gcp":"GCP",
    "oracle":"Oracle",
    "agile":"Agile"
})
```

```python
remove_skills = [
    "Application development",
    "Software development",
    "Development methodologies",
    "Computer science",
    "Project Management",
    "Backend",
    "Debugging"
]

skills = skills[
    ~skills.isin(remove_skills)
]
```

```python
top_skills = skills.value_counts()
```
